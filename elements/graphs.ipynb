```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

# Données pour les scénarios
temps = range(13)  # 12 mois + remboursement initial
depense_initiale = 656725
gains_pessimiste = [-depense_initiale, 886881.25 - depense_initiale]
gains_realiste = [-depense_initiale, 1886881.25 - depense_initiale]
gains_optimiste = [-depense_initiale, 2886881.25 - depense_initiale]

# Tracer le graphique
plt.plot(temps, [0] * len(temps), '--', color='#f33', label='Seuil de Rentabilité')
plt.plot([0, 12], gains_pessimiste, color='#ff3', label='Scénario Pessimiste (+ 1%)')
plt.plot([0, 12], gains_realiste, color='#38f', label='Scénario Réaliste (+ 2%)')
plt.plot([0, 12], gains_optimiste, color='#3f3', label='Scénario Optimiste (+ 3%)')

# Ajouter des titres et des étiquettes
plt.title('Scénarios d\'Atteinte de Rentabilité')
plt.xlabel('Temps (mois)')
plt.ylabel('Montant (M€)')
plt.legend()

# Afficher le graphique
plt.grid(True)
plt.show()
```

```python
# Données pour le burndown chart
sprints = [0, 2, 4, 6, 8]  # Sprints en semaines (0 pour le début)
temps_initial = 118.75  # Temps initial en heures
temps_restants = [118.75, 90, 55, 18, 0]  # Temps restant à la fin de chaque sprint

# Tracer le graphique
plt.plot(sprints, temps_restants, marker='o', color='red', label='Temps restant')

# Ajouter une ligne idéale de burndown
ideal_burndown = [temps_initial - (temps_initial / 8) * x for x in sprints]
plt.plot(sprints, ideal_burndown, 'b--', label='Ligne idéale')

# Ajouter des titres et des étiquettes
plt.title('Burndown Chart')
plt.xlabel('Temps (semaines)')
plt.ylabel('Temps restant (heures)')
plt.legend()

# Ajouter une grille
plt.grid(True)

# Afficher le graphique
plt.show()

```
